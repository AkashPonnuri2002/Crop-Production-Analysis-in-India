Crop Production Analysis in India

```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading Dataset
cp_data = pd.read_csv('Crop Production data.csv')

# Display the first five rows of the data
cp_data.head()
```

Handling Missing Data

```python
# Removing empty rows
cp_data.dropna()

# Checking for missing values
missing_values = cp_data.isnull().sum()
print('Missing values:\n')
print(missing_values)
```

```python
# Summary Statistics of Data
cp_data.describe()
```

Data Visualization

```python
# Plot crop production over the years

plt.figure(figsize=(12, 6))
sns.lineplot(x='Crop_Year', y='Production', data=cp_data)
plt.title('Crop Production Over the Years')
plt.xlabel('Year')
plt.ylabel('Production')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plot crop production by crop

plt.figure(figsize=(12, 6))
sns.barplot(x='Crop', y='Production', data=cp_data, estimator=sum)
plt.title('Crop Production by Crop')
plt.xlabel('Crop')
plt.ylabel('Production')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Plot crop production by state

plt.figure(figsize=(12, 6))
sns.barplot(x='State_Name', y='Production', data=cp_data, estimator=sum)
plt.title('Crop Production by State')
plt.xlabel('State')
plt.ylabel('Production')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Plot crop production by season

plt.figure(figsize=(12, 6))
sns.barplot(x='Season', y='Production', data=cp_data, estimator=sum)
plt.title('Crop Production by Season')
plt.xlabel('Season')
plt.ylabel('Production')
plt.tight_layout()
plt.show()
```

```python
# Plot crop production distribution

plt.figure(figsize=(12, 6))
sns.histplot(cp_data['Production'], bins=30, kde=True, color='orange', alpha=0.7)
plt.title('Crop Production Distribution')
plt.xlabel('Production')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
